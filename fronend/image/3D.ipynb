```python
import imageio
import os
import numpy as np
from pathlib import Path
from PIL import Image
from pygifsicle import optimize # 引入pygifsicle

def png_to_gif_optimized(path, output="result.gif", duration=0.5):
    """
    将指定目录下的PNG图片转换为高质量的GIF动画（背景透明）。
    使用Pillow和imageio生成，并调用Gifsicle进行最终优化。

    参数:
        path: PNG图片所在目录
        output: 输出GIF文件名
        duration: 每帧持续时间(秒)
    """
    try:
        path = Path(path)
        if not path.is_dir():
            raise NotADirectoryError(f"提供的路径不是一个目录: {path}")

        png_files = sorted(path.glob("*.png"))
        if not png_files:
            print("警告: 目录中未找到任何PNG文件。")
            return

        frames = []
        standard_size = None

        print("开始加载并处理PNG文件...")
        for file in png_files:
            try:
                with Image.open(file) as img:
                    if standard_size is None:
                        standard_size = img.size
                        print(f"检测到标准尺寸: {standard_size}")
                    
                    if img.size != standard_size:
                        print(f"警告: {file.name} 的尺寸 ({img.size}) 与标准尺寸不符，将进行调整。")
                        img = img.resize(standard_size, Image.LANCZOS)

                    # 确保是RGBA模式以支持透明度
                    if img.mode != 'RGBA':
                        img = img.convert('RGBA')
                    
                    frames.append(np.array(img))
                    print(f"  - 已加载: {file.name}")
            
            except Exception as e:
                print(f"读取文件 {file.name} 时发生错误: {e}")

        if not frames:
            print("错误: 没有成功加载任何有效的图片帧。")
            return

        output_path = Path(output)
        
        # 步骤1: 使用 imageio 生成初始GIF
        print(f"\n正在生成初始GIF文件到: {output_path}...")
        imageio.mimsave(
            output_path,
            frames,
            format='GIF',
            duration=duration,
            loop=0,
            disposal=2
        )
        print("初始GIF生成完毕。")
        
        # 步骤2: 使用 Gifsicle 优化GIF
        print("正在使用 Gifsicle 进行优化（可能会花费一些时间）...")
        try:
            # Gifsicle的优化会覆盖原文件，所以我们指定同一个输出路径
            optimize(str(output_path))
            print(f"优化完成！最终GIF已保存至: {output_path.absolute()}")
        except Exception as e:
            # 如果Gifsicle未安装或出错，会在这里捕获异常
            print(f"Gifsicle优化失败: {e}")
            print("请确保您已正确安装Gifsicle并将其添加到了系统PATH中。")
            print(f"未优化的GIF文件仍然保存在: {output_path.absolute()}")

    except Exception as e:
        print(f"处理过程中发生严重错误: {e}")

def main():
    # 替换为您的实际图片路径
    test_path = r"E:\1、heda\ogt\OGT OGA\IC\descriptor\web_test2 copy\fronend\image\ogt" 
    png_to_gif_optimized(
        test_path, 
        output="my_high_quality_result.gif", 
        duration=0.5
    )

if __name__ == "__main__":
    main()
```

```python
import imageio
import numpy as np
from pathlib import Path
from PIL import Image

def png_to_mp4(path, output="result.mp4", fps=2): # fps = 1/duration
    """
    将指定目录下的PNG图片转换为MP4视频（支持高质量和透明度）。

    参数:
        path: PNG图片所在目录
        output: 输出MP4文件名
        fps: 帧率 (frames per second)，每秒显示的帧数
    """
    try:
        path = Path(path)
        if not path.is_dir():
            raise NotADirectoryError(f"提供的路径不是一个目录: {path}")

        png_files = sorted(path.glob("*.png"))
        if not png_files:
            print("警告: 目录中未找到任何PNG文件。")
            return

        frames = []
        standard_size = None

        print("开始加载并处理PNG文件...")
        for file in png_files:
            try:
                with Image.open(file) as img:
                    if standard_size is None:
                        standard_size = img.size
                        print(f"检测到标准尺寸: {standard_size}")
                    
                    if img.size != standard_size:
                        print(f"警告: {file.name} 的尺寸 ({img.size}) 与标准尺寸不符，将进行调整。")
                        img = img.resize(standard_size, Image.LANCZOS)

                    # MP4通常需要RGB格式，透明度需要特定编码器支持
                    # 如果要支持透明度，需要使用RGBA并指定编码器，例如 'libvpx-vp9' 用于WebM
                    # 但对于大多数情况，如果背景是纯色，RGB就足够了
                    if img.mode == 'RGBA':
                        # 如果是RGBA，可以将其转换为RGB，并用白色背景填充透明区域
                        # 或者如果你有一个视频播放器支持透明度（如带alpha通道的WebM）
                        # 这里为了简单，先转成RGB，假设背景是纯黑或纯白
                        background = Image.new('RGB', img.size, (0, 0, 0)) # 假设黑色背景
                        background.paste(img, mask=img.split()[3]) # 使用alpha通道作为掩码
                        img = background
                    elif img.mode != 'RGB':
                        img = img.convert('RGB')
                    
                    frames.append(np.array(img))
                    print(f"  - 已加载: {file.name}")
            
            except Exception as e:
                print(f"读取文件 {file.name} 时发生错误: {e}")

        if not frames:
            print("错误: 没有成功加载任何有效的图片帧。")
            return

        output_path = Path(output)
        
        print(f"\n正在生成MP4视频文件到: {output_path}...")
        imageio.mimsave(
            output_path,
            frames,
            fps=fps, # 帧率
            quality=8 # 0-10，更高值表示更高质量，文件更大
        )
        print(f"MP4视频生成完毕。文件保存在: {output_path.absolute()}")

    except Exception as e:
        print(f"处理过程中发生严重错误: {e}")

def main():
    # 替换为您的实际图片路径
    test_path = r"E:\1、heda\ogt\OGT OGA\IC\descriptor\web_test2 copy\fronend\image\ogt" 
    png_to_mp4(
        test_path, 
        output="my_high_quality_video.mp4", 
        fps=2 # 对应 GIF 的 duration=0.5 (1/0.5 = 2 fps)
    )

if __name__ == "__main__":
    main()
```
